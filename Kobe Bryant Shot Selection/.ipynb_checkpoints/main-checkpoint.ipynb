```python
#drop rows with na values in the target feature and reset the index so we dont have anything missing
data = data[data['shot_made_flag'].notnull()].reset_index() 
```

```python
data['game_date'] = pd.to_datetime(data['game_date'])
data['game_day'] = data['game_date'].apply(lambda x: x.weekday()) ## 0 is Monday, 1 is Tuesday .... 7 is Sunday
```

I am going to use "minutes_remaining", "seconds_remaining" and "period" to create some additional features to have an easier grasp of the time

```python
data['period'].unique()
```

```python
# 12 minutes per period but it is a countdown so we subtract from 11
data['minutes_from_period_start'] = 11 - data['minutes_remaining']
data['seconds_from_period_start'] = 60*data['minutes_from_period_start'] + (60 - data['seconds_remaining']) 
# overtime lasts 5 minutes
data['minutes_from_game_start'] = (data['period'] <= 4).astype(int)*((data['period'] - 1 )*12 + data['minutes_from_period_start']) \
+ (data['period'] > 4).astype(int)*((data['period'] - 5)*5 + 4*12 + data['minutes_from_period_start'])
data['seconds_from_game_start'] = (data['period'] <= 4).astype(int)*((data['period'] - 1 )*12*60 + data['seconds_from_period_start']) \
+ (data['period'] > 4).astype(int)*((data['period'] - 5)*5*60 + 4*12*60 + data['minutes_from_period_start'])
```

```python
#sample and check if the new features added are correct ( compare 'period' with 'minutes_from_*_start' )
data[['period','minutes_remaining','seconds_remaining','minutes_from_period_start','seconds_from_period_start','minutes_from_game_start','seconds_from_game_start']].head()
```

### Let's answer the first question.
#### 1) What is Kobe's FG% and how many shots did he take?  


```python
sns.set(style='ticks')
data['shot_made_flag'] = data['shot_made_flag'].astype(int)
#percentage = [ (made shots) / (total shots) ] * 100
made_shots_num = data['shot_made_flag'].value_counts()[1]
total_shots = data['shot_made_flag'].value_counts()[0]+ data['shot_made_flag'].value_counts()[1]
fg_percentage = 100 * made_shots_num / total_shots
```

```python
sns.countplot(data=data,x='shot_made_flag',hue='shot_made_flag', palette=['r','g'])
plt.legend(labels=['miss', 'make'])
print('Kobe made {0} of his {1} shots.'.format(made_shots_num,total_shots) +\
'\nSo he ended his career with {:.3}% FG percentage.'.format(fg_percentage ) )
```

### Let's answer the second question.
#### 2) Can we plot Kobe's shot attempts and accuracy as the game progresses?  

```python
time_bins = np.arange(0,60*(4*12+3*5),20)+0.01 #20 means that every 20 seconds there is a bin
#attempts over time
attempts, b = np.histogram(data['seconds_from_game_start'],bins=time_bins)
#accuracay over time
made, b = np.histogram(data[data['shot_made_flag'] == 1]['seconds_from_game_start'],bins=time_bins)
miss, b = np.histogram(data[data['shot_made_flag'] == 0]['seconds_from_game_start'],bins=time_bins)
```

```python
acc = made.astype(float) / attempts
```

The error occurs because I am dividing by zero but that does not matter in the end the only problem is that I will not have correct accuracy data for my accuracy(time) plot <strong>ONLY on overtime periods</strong>

```python
fig, (ax1,ax2) = plt.subplots(2,1, figsize=(10,10))
ax1.plot(time_bins[1:],attempts)
maxHeight = max(attempts)+10
#draw red line at the point which each period ends
ax1.vlines(x=[0,12*60,2*12*60,3*12*60,4*12*60,4*12*60+5*60,4*12*60+2*5*60,4*12*60+3*5*60], ymin=0,ymax=maxHeight, colors='r', alpha=0.5)
ax1.hlines(y=[50,100,150,200,250,300,350,400], xmin=0, xmax=time_bins[-1] ,colors='b', alpha=0.1)
ax1.set_xlabel('time[sec]')
ax1.set_ylabel('shot attempts')
ax1.set_title('based on 20 second time bins')
#Let's plot his accuracy over time
ax2.plot(time_bins[1:], acc)
ax2.vlines(x=[0,12*60,2*12*60,3*12*60,4*12*60,4*12*60+5*60,4*12*60+2*5*60,4*12*60+3*5*60], ymin=0.29,ymax=0.6, colors='r', alpha=0.5)
ax2.hlines(y=[0.3,0.35,0.4,0.45,0.5,0.55,0.6], xmin=0, xmax=time_bins[-1] ,colors='b', alpha=0.1)
ax2.set_xlabel('time[sec]')
ax2.set_ylabel('accuracy')
ax2.set_title('based on 20 second time bins')
plt.tight_layout()
```

As we can see Kobe was trusted to take the last shot of every period and the first shot at every overtime period throughout his 20 year career  
He also seems to have trouble with his FG% at the end of the 4th period.  
Perhaps because presure to win the game gets to him?

Quite noteworthy that all those game winning end of period shots Kobe has on his highlights are pretty much because he took a lot of them rather than made most of them

### Continuing on time for the third question.
#### 3) Can we visualize spacially if distance from basket, court zone, shot type etc. makes a difference in shot accuracy?

```python
#hit and miss data that I will be using from now on
h = data[data['shot_made_flag'] == 1]
m = data[data['shot_made_flag'] == 0]
```

```python
court_scale, alpha = 7, 0.1
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(2 * court_scale, court_scale*(94.0/50.0))) # length 94ft width 50ft
# made shots ( h as in 'hit')
sns.scatterplot('loc_x', 'loc_y',data=h, color='g', alpha=alpha, ax=ax1)
ax1.set_title('Shots Made')
ax1.set_ylim([-50, 900])
# missed
sns.scatterplot('loc_x', 'loc_y',data=m, color='r', alpha=alpha, ax=ax2)
ax2.set_title('Shots missed')
ax2.set_ylim([-50, 900])
plt.tight_layout()
```

I used a small alpha value to be able to clearly see the areas where lots of "stacked" shots were made or missed.  
Seems like he missed a lot of shots underneath the basket which could mostly be due to fouls as noted from the strong red dot.  

Now time to use this 'spacial' plot in conjuction with data from other features and see if we can gather more precise info  
To do this I need to create some helper functions  

#### Helper Functions

```python
def zone_plot(feat):
    ''' Create a version of the plot shown above but 
    with different zones based on the input feature'''
    sns.set(style='ticks')
    alpha = 0.05
    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(2 * court_scale, court_scale*(94.0/50.0))) # length 94ft width 50ft
    #made
    sns.scatterplot('loc_x', 'loc_y',data=h.sort_values(by=feat), alpha=alpha, ax=ax1,hue=feat, palette='Set1')
    ax1.set_title('Shots Made')
    ax1.set_ylim([-50, 900])
    # missed
    sns.scatterplot('loc_x', 'loc_y',data=m.sort_values(by=feat), alpha=alpha, ax=ax2,hue=feat, palette='Set1')
    ax2.set_title('Shots missed')
    ax2.set_ylim([-50, 900])
    plt.tight_layout()
    return
```

```python
def acc(feat,flag):
    '''Print out the accuracy of Kobe's  shots and 
    if need be (indicated by the flag) return some 
    parameters used in my acc_plot functio'''   
    points = []
    names = []
    hit_sort = h.sort_values(feat,ascending=True)
    miss_sort = m.sort_values(feat,ascending=True)
    data_sort = data.sort_values(feat,ascending=True)
    for num, i in enumerate(data_sort[feat].unique()):
        total_i = len(hit_sort[hit_sort[feat]  == i].index) + len(miss_sort[miss_sort[feat]  == i].index)
        acc_i = ( len(hit_sort[hit_sort[feat] == i].index) ) / ( total_i ) 
        points.append(acc_i)
        names.append(i)
        if acc_i == 0:
            pass
        else:
            print(i,':','{:.3}%'.format(acc_i*100))
    if flag==0:
        return
    elif flag ==1:
        return names,num+1,points
```

```python
def acc_plot(names,num,points,width,height):
    ''' Create a plot with the y-axis being accuracy 
    and x-axis indicated by whichever feature I used 
    in my acc function that gave out the parameters 
    used in this one'''
    sns.set(style='whitegrid');
    plt.figure(figsize=(width,height));
    plt.plot(np.arange(0,num),points);
    plt.xticks(np.arange(0,num), names);
    plt.tight_layout;
    return
```
